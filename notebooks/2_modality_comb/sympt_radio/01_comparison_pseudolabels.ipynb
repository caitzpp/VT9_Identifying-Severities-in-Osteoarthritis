```python
import datetime
import os
import numpy as np
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
import config
from pathlib import Path
```

```python
today = datetime.date.today()

base_dir = config.RAW_DATA_PATH
proc_dir = config.PROC_DATA_PATH

rawimg_folder = "2025-09-13_hdbscan_img"
rawimg_folder_date = rawimg_folder.split('_')[0]
rawimg_run = "run32"

feature_folder = "2025-08-11_hdbscan"
feature_folder_date = feature_folder.split('_')[0]
feature_run = "run22"

rawimg_filepath =  os.path.join(proc_dir, "radiographic_features", rawimg_folder, rawimg_run)
feature_filepath =  os.path.join(proc_dir, "radiographic_features", feature_folder, feature_run)

ques_folder = "2025-08-23_hdbscan"
ques_folder_date = ques_folder.split('_')[0]
ques_run = "run2"

ques_filepath = os.path.join(proc_dir, ques_folder, "questionnaire", ques_run)


# load raw data
# file_name = 'questionnaires_raw.csv'

df = pd.read_csv(os.path.join(base_dir, 'brul_knee_annotations.csv'))

display(df.head())
```

```python
d = {"rawimg": rawimg_filepath,
     "features": feature_filepath,
     "questionnaires": ques_filepath}

for key, val in d.items():
    path_temp = d[key]
    for files in os.listdir(path_temp):
        if files.endswith("_scaled.csv"):
            df_name = files
            print(df_name)
            break
    df_temp = pd.read_csv(os.path.join(val, df_name))
    df = df.merge(df_temp, left_on="name",right_on = 'id', how="left", suffixes = ("", f"_{key}"))
```
